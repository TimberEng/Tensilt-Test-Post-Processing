```python
def read_lvm_raw(filepath):
    header = []
    data_lines = []
    in_data = False
    
    with open(filepath, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            if not in_data:
                header.append(line.strip())
                if "End_of_Header" in line:
                    in_data = True
            else:
                if line.strip():
                    data_lines.append(line.strip())
    
    # 拆分数据行（制表符或空格）
    data = [row.replace("\t", " ").split() for row in data_lines]
    return header, data

header, data = read_lvm_raw("6502.lvm")

print("Header preview:")
print("\n".join(header[:10]))
print("First 5 data rows:")
for row in data[:5]:
    print(row)
```
